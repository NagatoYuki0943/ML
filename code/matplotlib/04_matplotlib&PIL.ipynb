```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

# 绘制已有图片(与PIL联动)

```python
from PIL import Image
```

```python
image = Image.open('./img/71479714_p0.jpg')
```

## PIL在外部显示

```python
image.show()
```

## plt显示

```python
plt.figure(figsize=(10,8))
plt.axis('off') # 不显示坐标轴
plt.title('黄前久美子')
# vmin, vmax 代表图片亮度显示范围
plt.imshow(image, vmin=0, vmax=255)
# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['KaiTi']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
plt.show()
```

## 得到图片信息

```python
print(type(image))
print(image.size)  				#图片的尺寸
print(image.mode)  				#图片的模式
print(image.format)  			#图片的格式
print(image.getpixel((0,0)))	#得到像素：
```

## image.getpixel((宽, 高))

```python
print(image.getpixel((0,0)))	#得到像素
```

# 将 PIL Image 图片转换为 numpy 数组 np.array(image) [高, 宽,通道]

```python
img = np.array(image)
img
```

## numpy image 查看图片信息 .shape .dtypes

```python
print(img.shape)
print(img.dtype)
```

## numpy数据可以直接被画图 通道为 [高, 宽, 通道]

```python
plt.figure(figsize=(9, 16))
plt.imshow(img)
plt.axis('off')
plt.show()
```

## 将 numpy 数组转换为 PIL 图片  Image.fromarray(数组)

```python
new_img = Image.fromarray(img)

plt.figure(figsize=(9, 16))
plt.imshow(new_img)
plt.axis('off')
plt.show()
```
