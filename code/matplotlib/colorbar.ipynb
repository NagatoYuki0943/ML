```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.random.normal(0, 1, (100, 100))
x2 = np.random.uniform(0, 1, (100, 100))
```

```python
plt.imshow(x1)
plt.colorbar()
```

```python
plt.imshow(x2)
plt.colorbar()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 5))
for x, ax in zip([x1, x2], axes.flatten()):
    #  imshow 
    cax = ax.imshow(x)

    #  colorbar ()
    fig.colorbar(cax, ax=ax)
```
