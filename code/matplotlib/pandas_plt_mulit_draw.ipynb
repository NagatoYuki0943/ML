```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 读取数据

```python
df = pd.read_csv("Wofost72_WLP_FD.result.csv")
df
```

```python
df = df.set_index("day")
df.tail()
```

# pandas 绘制

```python
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(20, 32))
#                                 axes.flatten()
for column, ax in zip(df.columns, axes.flatten()):
    df[column].plot(ax=ax, title=column)
    ax.grid(True)
fig.autofmt_xdate()  # 下标变为斜体
```

# plt原生绘制,效果不好

```python
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(20, 32))
#                                 axes.flatten()
for column, ax in zip(df.columns, axes.flatten()):
    ax.plot(df.index, df[column], label=column)
    ax.grid(True)
fig.autofmt_xdate()  # 下标变为斜体
```

## 手动跳过数据展示更好看

```python
df["TAGP"][::20]
```

```python
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(20, 32))
#                                 axes.flatten()
for column, ax in zip(df.columns, axes.flatten()):
    ax.plot(df.index[::20], df[column][::20], label=column)
    ax.grid(True)
fig.autofmt_xdate()  # 下标变为斜体
```

```python
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(20, 32))
#                                 axes.flatten()
for column, ax in zip(df.columns, axes.flatten()):
    ax.plot_date(df.index[::20], df[column][::20], label=column)
    ax.grid(True)
fig.autofmt_xdate()  # 下标变为斜体
```

```python
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(20, 32))
#                                 axes.flatten()
for column, ax in zip(df.columns, axes.flatten()):
    #                                              g, k, g:, k: ...
    ax.plot_date(df.index[::20], df[column][::20], "g:", label=column)
    ax.grid(True)
fig.autofmt_xdate()  # 下标变为斜体
```
