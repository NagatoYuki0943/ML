```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Wofost72_WLP_FD.result.csv")
df
```

```python
df = df.set_index("day")
df.tail()
```

```python
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(16, 32))
#                                                axes.flatten()
for column, ax in zip(df.columns, axes.flatten()):
    df[column].plot(ax=ax, title="column")
    ax.grid(True)
fig.autofmt_xdate() # 下标变为斜体
```
