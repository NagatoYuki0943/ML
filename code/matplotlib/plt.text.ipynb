```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 0.5)
y = np.sin(x) + np.random.normal(0, 0.1, len(x))
```

```python
plt.figure(figsize=(20, 10))

plt.plot(x, y)

# plt.text ha va 参数
# ha（horizontalalignment）: 该参数决定文本在水平轴上的对齐方式。可选值包括：
#   - 'left': 文本左对齐。
#   - 'center': 文本居中对齐。
#   - 'right': 文本右对齐。
# va（verticalalignment）: 该参数决定文本在垂直轴上的对齐方式。可选值包括：
#   - 'top': 文本顶部对齐。
#   - 'center': 文本垂直居中对齐。
#   - 'bottom': 文本底部对齐。
#   - 'baseline': 文本基线对齐。
for _x, _y in zip(x, y):
    plt.text(_x, _y, f"{_y:.2f}", ha='center', va='top', fontsize=12)

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot([1, 2, 3], [1, 4, 9])

# plt.text 其他参数
# fontdict
#     - 一个字典，用于覆盖文本的默认属性，如字体、大小和颜色等。
#     - 如果fontdict为None，则使用默认值。
# fontsize
#     - 文本的字体大小。
#     - 可以是数字，表示绝对大小，或者是字符串，表示相对大小，如'xx-small'、'x-large'等。
# color 或 c
#     - 文本的颜色。
#     - 可以是颜色名称、十六进制代码、RGB或RGBA元组等。
# rotation 或 angle
#     - 文本的旋转角度。
#     - 正值为逆时针旋转。
# bbox
#     - 一个字典，用于设置文本周围的边框。
#     - 常用参数包括boxstyle（边框样式）、facecolor（填充色）、edgecolor（边框色）和edgewidth（边框宽度）等。
# transform
#     - 用于指定文本坐标系的变换。
#     - 默认为数据坐标系，可以设置为其他坐标系，如figure、axes等。
# clip_on
#     - 是否裁剪文本超出坐标轴范围的部分。
#     - 默认为False，即不裁剪。
plt.text(
    2,
    6,
    r'$\mathcal{A} \to \mathcal{B}$',
    fontdict={'color': 'r', 'size': 16},
    rotation=45,
    bbox=dict(boxstyle='round', facecolor='wheat', ec='k', lw=2)
)

plt.show()
```
