```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 0.5)
y = np.sin(x) + np.random.normal(0, 0.1, len(x))
```

```python
plt.figure(figsize=(20, 10))

plt.plot(x, y)

# plt.text ha va 
# hahorizontalalignment: 
#   - 'left': 
#   - 'center': 
#   - 'right': 
# vaverticalalignment: 
#   - 'top': 
#   - 'center': 
#   - 'bottom': 
#   - 'baseline': 
for _x, _y in zip(x, y):
    plt.text(_x, _y, f"{_y:.2f}", ha='center', va='top', fontsize=12)

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot([1, 2, 3], [1, 4, 9])

# plt.text 
# fontdict
#     - 
#     - fontdictNone
# fontsize
#     - 
#     - 'xx-small''x-large'
# color  c
#     - 
#     - RGBRGBA
# rotation  angle
#     - 
#     - 
# bbox
#     - 
#     - boxstylefacecoloredgecoloredgewidth
# transform
#     - 
#     - figureaxes
# clip_on
#     - 
#     - False
plt.text(
    2,
    6,
    r'$\mathcal{A} \to \mathcal{B}$',
    fontdict={'color': 'r', 'size': 16},
    rotation=45,
    bbox=dict(boxstyle='round', facecolor='wheat', ec='k', lw=2)
)

plt.show()
```
