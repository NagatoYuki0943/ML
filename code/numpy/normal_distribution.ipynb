```python
import numpy as np
import matplotlib.pyplot as plt
```

# σ的作用

- 横轴区间（μ-σ,μ+σ）内的面积为68.268949%
- 横轴区间（μ-2σ,μ+2σ）内的面积为95.449974%
- 横轴区间（μ-3σ,μ+3σ）内的面积为99.730020%

## full

```python
x = np.random.normal(loc=0, scale=1, size=1_000_000)
```

```python
plt.hist(x, bins=1000)
plt.show()
```

## σ=0.5

```python
x0_5 = x[np.bitwise_and(x>(-0.5), x<0.5)]
print(x0_5.shape)
plt.hist(x0_5, bins=1000)
plt.show()
```

## σ=1

```python
x1 = x[np.bitwise_and(x>(-1), x<1)]
print(x1.shape)
plt.hist(x1, bins=1000)
plt.show()
```

## σ=1.5

```python
x1_5 = x[np.bitwise_and(x>(-1.5), x<1.5)]
print(x1_5.shape)
plt.hist(x1_5, bins=1000)
plt.show()
```

## σ=2

```python
x2 = x[np.bitwise_and(x>(-2), x<2)]
print(x2.shape)
plt.hist(x2, bins=1000)
plt.show()
```

## σ=2.5

```python
x2_5 = x[np.bitwise_and(x>(-2.5), x<2.5)]
print(x2_5.shape)
plt.hist(x2_5, bins=1000)
plt.show()
```

## σ=3

```python
x3 = x[np.bitwise_and(x>(-3), x<3)]
print(x3.shape)
plt.hist(x3, bins=1000)
plt.show()
```

# compare

```python
a = [0.5, 1, 1.5, 2, 2.5, 3]
b = [x0_5.shape[0], x1.shape[0], x1_5.shape[0], x2.shape[0], x2_5.shape[0], x3.shape[0]]
b = [b_/1_000_000 for b_ in b]
plt.plot(a, b)
plt.grid()
for a_, b_ in zip(a, b):
    plt.text(a_-0.1, b_+0.01, f"{b_:.3f}", c='red')
```
