```python
import numpy as np
import matplotlib.pyplot as plt
```

# 一维数据

```python
x = np.array([1, 2, 3, 5, 7, 10, 14, 19, 25])
x
```

```python
grad_x = np.gradient(x)
grad_x
```

# 二维数据，图像

```python
# 创建一个简单的图像，这里使用一个带有明显边缘的图像
x = np.zeros((100, 100))
x[30:70, 30:70] = 1
plt.imshow(x)
```

```python
grad_x, grad_y = np.gradient(x)
grad_x.shape, grad_y.shape
```

```python
np.unique(grad_x), np.unique(grad_y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
axes[0].imshow(x)
axes[0].set_title('x')
axes[1].imshow(grad_x)
axes[1].set_title('grad_x')
axes[2].imshow(grad_y)
axes[2].set_title('grad_y')
```

# 三维数据，图像

```python
# 创建一个简单的图像，这里使用一个带有明显边缘的图像
x = np.zeros((100, 100, 3))
x[:, :, 0] = 0
x[:, :, 1] = 1
x[:, :, 2] = 2
x[30:70, 30:70] = 3
plt.imshow(x)
```

```python
grad_x, grad_y, grad_z = np.gradient(x)
grad_x.shape, grad_y.shape, grad_z.shape
```

```python
np.unique(grad_x), np.unique(grad_y), np.unique(grad_z)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
axes[0].imshow(x)
axes[0].set_title('x')
axes[1].imshow(grad_x)
axes[1].set_title('grad_x')
axes[2].imshow(grad_y)
axes[2].set_title('grad_y')
axes[3].imshow(grad_z)
axes[3].set_title('grad_z')
```
