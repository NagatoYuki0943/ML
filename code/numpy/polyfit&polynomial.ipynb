# polyfit简单线性回归

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 创建数据
weight = np.array([1.5, 2.3, 3.7, 2.2, 2.9, 4.1, 3.8, 2.6, 3.4, 4.6])
fuel_efficiency = np.array([38, 37, 32, 36, 34, 31, 32, 35, 33, 30])
```

```python
w, b = np.polyfit(x=weight, y=fuel_efficiency, deg=1)
w, b
```

```python
# 创建散点图
plt.figure(figsize=(8, 6))
plt.scatter(weight, fuel_efficiency)
plt.plot(weight, w * weight + b, color='red')
plt.title('Scatter plot of weight vs fuel efficiency')
plt.xlabel('Weight')
plt.ylabel('Fuel Efficiency')
plt.grid(True)
```

# polynomial 新api

https://numpy.org/doc/stable/reference/generated/numpy.polynomial.polynomial.polyfit.html

```python
b, w = np.polynomial.polynomial.polyfit(x=weight, y=fuel_efficiency, deg=1)
w, b
```

```python
# 创建散点图
plt.figure(figsize=(8, 6))
plt.scatter(weight, fuel_efficiency)
plt.plot(weight, w * weight + b, color='red')
plt.title('Scatter plot of weight vs fuel efficiency')
plt.xlabel('Weight')
plt.ylabel('Fuel Efficiency')
plt.grid(True)
```
