```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# standard = (x - mean) / std

```python
x = np.random.normal(20, 100, 10000)
x = x.astype(np.float32)
x[:10]
```

```python
sns.kdeplot(x)
plt.grid()
```

```python
sns.kdeplot((x - x.mean()) / x.std())
plt.grid()
```
