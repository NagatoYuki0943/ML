# 最小二乘法

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 接受两组数据x和y作为输入，并计算出最佳拟合直线的斜率b1和截距b0。
def least_squares(x, y):
    # x和y是观测数据点
    # 计算x和y的均值
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    # 计算斜率b1和截距b0
    b1 = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)
    b0 = y_mean - b1 * x_mean

    return b0, b1
```

```python
# 示例数据点
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([1, 2, 1.3, 3.75, 3.5, 5.1])
```

```python
# 计算最佳拟合线
b0, b1 = least_squares(x, y)
# 打印结果
print(f"最佳拟合线的截距是: {b0}")
print(f"最佳拟合线的斜率是: {b1}")
```

```python
# 画出原始数据点
plt.scatter(x, y, color='blue', label='Original data')

# 画出最佳拟合线
x_line = np.array([0, 6])  # 为了画出更长的线，这里扩展了x的范围
y_line = b1 * x_line + b0
plt.plot(x_line, y_line, color='red', label='Fitted line')

# 添加图例
plt.legend()

# 显示图表
plt.show()
```
