```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 添加新列

```python
data = pd.DataFrame(np.random.random((5, 2)), columns=["a", "b"])
data
```

## df["new_column"] = new_data 添加新列

```python
data["c"] = np.random.random(5)
data
```

## df.loc[:, "new_column"] = new_data 添加新列

可以参考 2_商品评论情感分析_MultinomialNB.ipynb

```python
# 在里面添加列才有效
data.loc[:, "d"] = np.random.random((5))
data
```

```python
# loc[:]这样使用无效
data.loc[:]["e"] = np.random.random((5))
data
```

```python
# [:, :]["e"]这样使用有效,等价data["e"]
data.loc[:, :]["e"] = np.random.random((5))
data
```

## pd.insert(loc=第几列, column="列名", value=值) 插入列

```python
data.insert(5, column="f", value=np.random.randint(1, 10, 5))
data
```

# 删除列或者行 df.drop
- labels 单个标签或者标签列表
- axis=0 默认 删除index
- axis=1 指定删除列
- inplace=True 修改原数据
- level 针对多重索引 指定级别
- index 指定索引
- columns 指定列名

## 删除行 pd.drop(axis=0, index=行], inplace=False)

```python
data.drop(axis=0, index=1, inplace=False)
```

```python
data.drop(axis=0, index=[1, 4], inplace=False)
```

## pd.drop(axis=1, columns=列名, inplace=False)

```python
data.drop(axis=1, columns="a", inplace=False)
```

```python
data.drop(axis=1, columns=["a", "b"], inplace=False)
```

# 修改columns

```python
data1 = pd.DataFrame(np.random.random((5, 5)), columns=["a", "b", "c", "d", "e"])
data1
```

## df.columns = new_columns

```python
data1.columns = ["语文", "数学", "英语", "政治", "体育"]
data1
```

# 修改index

## df.index = new_index 必须整体全部修改

```python
stu = ["学生_" + str(i + 1) for i in range(data.shape[0])]
stu
```

```python
data1.index = stu
data1
```

```python
# 单个值不能设置
data1.index[0] = "学生1"
```

## 重设索引 reset_index(drop=False, inplace=False)
reset_index(drop=False, inplace=False)

- 默认添加一列,从0开始
- drop : boolean
    - default False添加一列,从0开始
    - True 当做新的索引，删除原来的索引
- inplace: False 是否修改原值

```python
data1.reset_index()
```

```python
data1.reset_index(drop=True)
```

## 以某列值设置为新的索引 set_index()
set_index(keys, drop=True, inplace=False)

- 默认添加一列,从0开始
- keys : 列索引名成或者列索引名称的列表
- drop : boolean,
    - default True.当做新的索引，删除原来的列
- inplace: False 是否修改原值

```python
data1.set_index("语文")
```

```python
data1.set_index(["语文", "体育"])
```

# 修改值

```python
data2 = pd.DataFrame(np.random.randn(5, 5), columns=["a", "b", "c", "d", "e"])
data2
```

## df["column"] = new_value 修改整列

```python
data2["a"] = 10
data2
```

## df.column = new_value 修改整列

```python
data2.b = 20
data2
```

## 获取指定位置修改值

```python
data2[:2] = -100
data2
```

```python
data2.iloc[2:5, 1:3] = 0
data2
```

```python
data2.loc[3:5, "d":"e"] = 999
data2
```
