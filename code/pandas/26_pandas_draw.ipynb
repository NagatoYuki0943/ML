```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 准备数据

```python
# 生成10名同学，5门功课的数据
score = np.random.randint(40, 100, (10, 5))
score
```

```python
# 构造列索引序列
subjects = ["Chinese", "Math", "English", "Sciencs", "Physics"]
# 构造行索引序列
students = [f"stu_{i+1}" for i in range(score.shape[0])]
# 添加行列索引
score_pd = pd.DataFrame(score, columns=subjects, index=students)
score_pd
```

```python
# 读取文件
data = pd.read_csv("../data/stock_day.csv")
# 删除一些列，让数据更简单些，再去做后面的操作
data = data.drop(["ma5", "ma10", "ma20", "v_ma5", "v_ma10", "v_ma20"], axis=1)
data
```

```python
# 对p_change进行求和
stock_rise = data['p_change']
stock_rise.head()
```

# df.plot(kind)
a.plot (kind="line")

-   kind : str，需要绘制图形的种类
    -  line : 折线图 plot (default)
    -  bar : 条形图
    -  barh : 横向条形图
        -  关于“barh”的解释：
        -  http://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.barh.html
    -  hist : 直方图
    -  pie : 饼图
    -  scatter : 散点图

```python
score_pd.plot(kind='bar')
```
