# 高级处理-合并

```python
score = np.random.randint(40, 100, (5, 5))
# 构造列索引序列
subjects = ["Chinese", "Math", "English", "Sciencs", "Sports"]
# 构造行索引序列
students = [f"stu_{i+1}" for i in range(score.shape[0])]
# 添加行索引
score_df1 = pd.DataFrame(score, columns=subjects, index=students)
score_df1
```

```python
score = np.random.randint(40, 100, (5, 5))
# 构造列索引序列
subjects = ["Chinese", "Math", "English", "Sciencs", "Sports"]
# 构造行索引序列
students = [f"stu_{i+6}" for i in range(score.shape[0])]
# 添加行索引
score_df2 = pd.DataFrame(score, columns=subjects, index=students)
score_df2
```

## append (deprecated)

```python
score_df1.append(score_df2)
```

## concat([data...], axis=0)

axis = 0 index拼接

axis = 1 column拼接

```python
pd.concat([score_df1, score_df2], axis=0)
```

```python
# 这里可以看到会按照行和列进行对比拼接, 数据没有的部分就为NaN
pd.concat([score_df1, score_df2], axis=1)
```

## pd.merge
pd.merge(left, right, how='inner', on=None)

- 可以指定按照两组数据的共同键值对合并或者左右各自
-  left: DataFrame
-  right: 另一个DataFrame
-  on: 指定的共同键
- how: 按照什么方式连接

```python
# 创建两张表
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

### 内连接(默认)

```python
pd.merge(left, right, on=["key1"])
```

```python
#                       使用两个键进行匹配,都匹配才算
pd.merge(left, right, on=["key1", "key2"])
```

### 左连接

```python
pd.merge(left, right, on=["key1"], how="left")
```

```python
pd.merge(left, right, on=["key1", "key2"], how="left")
```

### 右连接

```python
pd.merge(left, right, on=["key1"], how="right")
```

```python
#                     使用两个键进行匹配,都匹配才算
pd.merge(left, right, on=["key1", "key2"], how="right")
```

### 外链接

```python
#                     使用两个键进行匹配,都匹配才算
pd.merge(left, right, on=["key1", "key2"], how="outer")
```
