```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.normal(0, 1, 10000)
sns.displot(x, bins=20)
```

```python
bins = np.arange(-4, 5)
bins
```

```python
histogram = np.histogram(x, bins=bins, density=True)[0]
```

```python
bins = 0.5 * (bins[1:] + bins[:-1])
bins
```

```python
pdf = stats.norm.pdf(bins)
pdf
```

```python
plt.plot(bins, histogram, label="histogram")
plt.plot(bins, pdf, label="pdf")
plt.grid(True)
plt.legend()
```
