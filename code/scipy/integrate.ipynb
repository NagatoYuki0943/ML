```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

# integrate 积分

# quad

- quadratic 二次的
- quadrant 象限
- quadrature 正交


## $I(f)=\int_{a}^{b} (x+1)dx$

```python
def f(x):
    return x + 1
```

```python
x = np.arange(0, 3, 0.1)
x
```

```python
y = f(x)
y
```

```python
plt.plot(x, y)
plt.grid(True)
```

```python
v, err = integrate.quad(f, a=0, b=1)
v, err
# 积分为1.5
```

## $I(f)=\int_{a}^{b} \frac 1 {\sqrt {\left| x \right|}}dx$

如果遇到积分函数有断点，可以通过quad函数的points给出断点继续求积分

```python
def f(x):
    return 1 / (np.sqrt(np.abs(x)))
```

```python
x = np.arange(-2, 2, 0.1)
x
```

```python
y = f(x)
y
```

```python
plt.plot(x, y)
plt.grid()
```

```python
integrate.quad(f, a=-1, b=1)
# 报错
```

```python
integrate.quad(f, a=-1, b=1, points=[0])
# 报错
```

# trapz 求点集的积分

```python
def f(x):
    return np.sqrt(x)
```

```python
x = np.linspace(0, 2, 20)
x
```

```python
y = f(x)
y
```

```python
v = integrate.trapz(y=y, x=x)
v
# 积分结果
```

# 多重积分

SciPy下的二重积分可以用dblquad函数来计算、三重积分可以用tplquad函数来计算而关于f(x1,x2,⋯,xn)的多重积分可以使用nquad函数

```python
def f(x, y):
    return x * y


def h(x):
    return x**2
```

```python
v, err = integrate.dblquad(f, a=1, b=2, gfun=lambda x: 1, hfun=h)
v, err
```
