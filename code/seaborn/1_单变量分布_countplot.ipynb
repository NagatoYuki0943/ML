# 导入包

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

## 一维数组

```python
arr1 = np.random.randint(1, 10, 1000)
arr1[:10]
```

### 参数1为data,如果是一维数据不分类

```python
sns.countplot(arr1)
```

### 传递给参数x是竖直显示

```python
sns.countplot(x=arr1)
```

### 传递给参数y是水平显示

```python
sns.countplot(y=arr1)
```

```python
sns.histplot(arr1)
```

## DataFrame数据

```python
arr2 = np.random.rand(3, 1000).reshape(1000, 3)
arr2
```

```python
data = np.concatenate((arr2, arr1.reshape(1000, -1)), axis=1)
data.shape
```

```python
df = pd.DataFrame(data=data, columns=["data1", "data2", "data3", "target"])
df
```

### 传递data和x

```python
sns.countplot(data=df, x="target")
```

### 只传递x

```python
sns.countplot(x=df["target"])
```

```python
sns.histplot(df, x="target", hue="target")
```
