```python
# 导入
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Seaborn----绘制统计图形

```python
# 生成100个随机数
# 随机数种子
np.random.seed(0)
arr = np.random.randn(100)
arr
```

## displot

```python
#            数量,用于控制条形的数量
sns.displot(arr, bins=10)
```

## histplot

```python
sns.histplot(arr, bins=10)
```

```python
sns.displot(arr, bins=10, kind="hist")
```

```python
sns.histplot(arr, bins=10, kde=True)
```

```python
sns.histplot(arr, bins=10, kde=True, stat="density")
```

## kdeplot

```python
sns.kdeplot(arr)
```

```python
sns.displot(arr, kind="kde")
```

## ecdfplot

```python
sns.ecdfplot(arr)
```

```python
sns.displot(arr, kind="ecdf")
```
