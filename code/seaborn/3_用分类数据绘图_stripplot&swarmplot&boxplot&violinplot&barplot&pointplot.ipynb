```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 用分类数据绘图

## 类别散点图
seaborn.stripplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, jitter=False)
- (1) x，y, hue:   data中的列名, x和y用来显示, hue用来改变颜色
- (2) data：用于绘制的数据集。如果x和y不存在，则它将作为宽格式，否则将作为长格式。
- (3) jitter：表示抖动的程度(仅沿类別轴)。当很多数据点重叠时，可以指定抖动的数量或者设为Tue使用默认值。

```python
# 获取tips数据
tips = sns.load_dataset("tips")
tips
```

### stripplot

```python
# 数据会重叠
sns.stripplot(data=tips, x="day", y="total_bill", hue="day")
```

```python
# 从上图中可以看出，图表中的横坐标是分类的数据，而且一些数据点会互相重叠，不易于观察。
# 为了解决这个问题，可以在调用striplot()函数时传入jitter参数，以调整横坐标的位置，改后的示例代码如下。
# 还是会有重叠
sns.stripplot(data=tips, x="day", y="total_bill", hue="day", jitter=True)
```

### stripplot 参数

```python
# hue: 显示不同
# 数据抖动
sns.stripplot(data=tips, x="day", y="total_bill", hue="time", jitter=True)
```

### swarmplot 所有的数据点都不会重叠
- x,y: data中的列名, x和y用来显示
- hue: 用来改变颜色

```python
# 除此之外，还可调用 swarmplot函数绘制散点图，该函数的好处是所有的数据点都不会重叠，
# 可以很清晰地观察到数据的分布情况，示例代码如下。
sns.swarmplot(data=tips, x="day", y="total_bill", hue="day")
```

## 类别内的数据分布

### 绘制箱形图 boxplot
seaborn.boxplot(x=None, y=None, hue=None, data=None, orient=None, color=None,  saturation=0.75, width=0.8)
- (1) x，y, hue:   data中的列名, x和y用来显示, hue用来改变颜色

- (2) palette：用于设置不同级别色相的颜色变量。---- palette=["r","g","b","y"]
- (3) saturation：用于设置数据显示的颜色饱和度。---- 使用小数表示

```python
sns.boxplot(data=tips, x="day", y="total_bill")
```

```python
# hue色调
sns.boxplot(data=tips, x="day", y="total_bill", hue="time")
```

```python
# palette 颜色
sns.boxplot(
    data=tips, x="day", y="total_bill", palette=["green", "blue", "yellow", "pink"]
)
```

```python
# saturation 饱和度
sns.boxplot(data=tips, x="day", y="total_bill", saturation=0.8)
```

### 绘制提琴图 violinplot
seaborn.violinplot(x=None, y=None, hue=None, data=None)
- x，y, hue:   data中的列名, x和y用来显示, hue用来改变颜色

```python
sns.violinplot(data=tips, x="day", y="total_bill")
```
