```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

# 载入数据

```python
iris = load_iris()
```

```python
iris.data[:10]
```

```python
iris.feature_names
```

```python
feature_names = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
```

```python
df = pd.DataFrame(iris.data, columns=feature_names)
df
```

# 多维数组多次画图

```python
# 使用数组数据label为0 1 2 3...
sns.histplot(iris.data)
```

```python
sns.kdeplot(iris.data)
```

# DataFrame画多维度图

```python
# 使用DataFrame或者字典数据会显示正确的label
sns.histplot(df)
```

```python
sns.kdeplot(df)
```

# 字典多次画图

```python
# 使用字典传值
sns.histplot(
    {
        "sepal_length": iris.data[:, 0],
        "sepal_width": iris.data[:, 1],
        "petal_length": iris.data[:, 2],
        "petal_width": iris.data[:, 3],
    }
)
```

```python
sns.kdeplot(
    {
        "sepal_length": iris.data[:, 0],
        "sepal_width": iris.data[:, 1],
        "petal_length": iris.data[:, 2],
        "petal_width": iris.data[:, 3],
    }
)
```
