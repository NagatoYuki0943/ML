# 导入包

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arr1 = np.random.randint(1, 10, 1000)
arr1[:10]
```

# 画图函数的 ax 参数可以传递 axes[0][1], 表示位置

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 12))
sns.countplot(arr1, ax=axes[0][0])
sns.countplot(x=arr1, ax=axes[0][1])
sns.countplot(y=arr1, ax=axes[1][0])
sns.histplot(arr1, ax=axes[1][1])
```
