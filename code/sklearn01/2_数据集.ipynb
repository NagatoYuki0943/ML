# sklearn获取数据集
sklearn小数据集    sklearn.datasets.load_*() <br>
sklearn大数据集    sklearn.datasets.fetch_*(data_home=None) <br>
sklearn.datasets.fetch_20newsgroups(data_home=None,subset=‘train’) <br>
    data_home: 路径 <br>
    subset：'train'或者'test'，'all'，可选，选择要加载的数据集。 <br>
    训练集的“训练”，测试集的“测试”，两者的“全部” <br>

# sklearn数据集返回值介绍
load和fetch返回的数据类型datasets.base.Bunch(字典格式) <br>
    data：特征数据数组，是 [n_samples * n_features] 的二维 numpy.ndarray 数组 <br>
    target：标签数组，是 n_samples 的一维 numpy.ndarray 数组 <br>
    DESCR：数据描述 <br>
    feature_names：特征名,新闻数据，手写数字、回归数据集没有 <br>
    target_names：标签名

# 把数据转换成dataframe的格式

```python
df = pd.DataFrame(
    data=iris.data,
    columns=["Sepal_Length", "Sepal_Width", "Petal_Length", "Petal_Width"],
)
df.head()
```

```python
# df动态添加列
df["Species"] = iris.target
df
```

# 画图

```python
sns.pairplot(df.iloc[:, :4])
```

```python
def iris_plot(col1, col2, fit_reg=False):
    sns.lmplot(
        data=df, x=col1, y=col2, hue="Species", fit_reg=fit_reg
    )  # fit_reg 是否显示拟合直线
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.title("鸢尾花种类分布图")

    # 解决中文显示问题
    plt.rcParams["font.sans-serif"] = ["KaiTi"]  # 指定默认字体
    plt.rcParams["axes.unicode_minus"] = False  # 解决保存图像是负号'-'显示为方块的问题
    plt.show()
```

```python
iris_plot("Sepal_Length", "Sepal_Width")
```

```python
iris_plot("Sepal_Length", "Sepal_Width", True)
```

```python
iris_plot("Petal_Length", "Petal_Width")
```

```python
iris_plot("Petal_Length", "Petal_Width", True)
```
