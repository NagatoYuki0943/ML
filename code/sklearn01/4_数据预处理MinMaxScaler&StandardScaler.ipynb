```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import seaborn as sns
```

```python
data = np.random.randint(0, 256, 10000)
data
```

# fit、transform、fit_transform
数据预处理中的使用:
- fit(): 求得训练集X的均值，方差，最大值，最小值,这些训练集X固有的属性
- transform(): 在fit的基础上，进行标准化，降维，归一化等操作
- fit_transform(): fit和transform的组合，既包括了训练又包含了转换

使用方法：
1. fit_transform(trainData) <br/>
    对trainData进行fit的整体指标，找到该part的整体指标，如均值、方差、最大值最小值等等（根据具体转换的目的）

2. transform(testData) <br/>
    对testData使用同样的均值、方差、最大最小值等指标进行转换，从而保证train、test处理方式相同

注意：
1. 必须先用fit_transform(trainData)，之后再transform(testData)，直接transform(testData)，程序会报错
2. 如果fit_transfrom(trainData)后，使用fit_transform(testData)而不transform(testData)，虽然也能归一化，但是两个结果不是在同一个“标准”下的，具有明显差异。


# MinMaxScaler (feature_range=(0,1)... ) 归一化处理
- feature_range=(0,1)    指的是归一化的数字范围,随便写

```python
min_max_scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
data_min_max_scaled = min_max_scaler.fit_transform(data.reshape(-1, 1))
data_min_max_scaled[:10]
```

```python
sns.kdeplot(data_min_max_scaled)
```

# StandardScaler
处理之后每列来说所有数据均值为0,标准差为1


```python
standard_scaler = StandardScaler()
```

```python
data_standard_scaled = standard_scaler.fit_transform(data.reshape(-1, 1))
data_standard_scaled[:10]
```

```python
sns.kdeplot(data_standard_scaled)
```
