```python
import numpy as np
import matplotlib.pyplot as plt
# 线性回归
from sklearn.linear_model import LinearRegression
# 决策树回归
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
```

# 1.生成数据

```python
x = np.arange(1, 11, 1)
x
```

```python
x = x.reshape(-1, 1)
x
```

```python
y = np.array([5.56, 5.70, 5.91, 6.40, 6.80, 7.05, 8.90, 8.70, 9.00, 9.05])
```

# 2.训练模型

```python
# 线性回归
linear_model = LinearRegression()
# 决策树
tree_1 = DecisionTreeRegressor(max_depth=1)
tree_3 = DecisionTreeRegressor(max_depth=3)
```

```python
linear_model.fit(x, y)
tree_1.fit(x, y)
tree_3.fit(x, y)
```

# 3.模型预测

```python
# 等距离等差数列
x_test = np.arange(0.0, 10.0, 0.01).reshape(-1, 1)
x_test[:10]
```

```python
linear_res = linear_model.predict(x_test)
tree_1_res = tree_1.predict(x_test)
tree_3_res = tree_3.predict(x_test)
```

# 4.结果可视化
可以看到线性模型不能很好的拟合非线性数据,决策树拟合的更好

```python
plt.figure(figsize=(10, 6), dpi=100)

# x,y代表原始值
# 点图
plt.scatter(x, y, label="origin", color="magenta")
# 折线图
plt.plot(x_test, linear_res, label="linear")
plt.plot(x_test, tree_1_res, label="tree_1")
plt.plot(x_test, tree_3_res, label="tree_3")

plt.legend()

plt.xlabel("data")
plt.ylabel("target")
plt.title("线性回归和回归决策树对比")

# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['KaiTi'] # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
# 显示图像
plt.show()
```
