```python
import numpy as np
import pandas as pd
import json
from pathlib import Path
import matplotlib.pyplot as plt
```

# load

```python
result_path = Path("./results/camera0_result.jsonl")
result_path.exists()
```

```python
results = []
with open(result_path, mode='r', encoding='utf-8') as f:
    for line in f:
        result = json.loads(line)
        results.append(result)
results[0]
```

# up

```python
up_thresholds = [result['threshold'] for result in results if 'up' in result['timestamp']]
up_center_xss = [result['center_xs'] for result in results if 'up' in result['timestamp']]
up_center_yss = [result['center_ys'] for result in results if 'up' in result['timestamp']]
up_center_xs = [result['center_x_mean'] for result in results if 'up' in result['timestamp']]
up_center_ys = [result['center_y_mean'] for result in results if 'up' in result['timestamp']]
```

```python
plt.figure(figsize=(20, 8))
plt.plot(up_thresholds)
plt.grid()
plt.title("up_threshold")
```

```python
plt.figure(figsize=(20, 8))
plt.plot(up_center_xs)
plt.grid()
plt.title("up_center_x")
```

```python
plt.figure(figsize=(20, 8))
plt.plot(up_center_ys)
plt.grid()
plt.title("up_center_y")
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(up_center_xs,up_center_ys, marker='.')
plt.grid()
plt.title("up_center")
```
